```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
X = iris['data']
y = iris['target']
target_names = iris['target_names']
columns = ['Sepal length','Sepal width','Petal length','Petal width']
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
df = pd.DataFrame(X)
df.columns = columns
df['target'] = y
```

3d contour plotting tutorial: https://scikit-learn.org/stable/auto_examples/svm/plot_iris_svc.html

```python
def solve(C,train_X,train_y,test_X,test_y,kernel,class1,class2,f1,f2,plot_fig):
    svc = SVC(kernel = kernel,C = C)
    svc.fit(train_X,train_y)
    pred = svc.predict(test_X)
    report = classification_report(test_y,pred,output_dict = True,target_names = [target_names[f1],target_names[f2]])
    report = pd.DataFrame(report).transpose()
    print("Using SVM with features "+class1+" and "+class2+ " ,kernel: "+kernel+ " and C  = "+str(C))
    display(report)
    if(plot_fig == False):
        return
    fnd = [0,0,0]
    for ind,color in zip([0,1,2],['r','g','b']):
        for i in range(len(train_X)):
            if(train_y[i] == ind):
                plt.scatter(train_X[i][0],train_X[i][1],color = color,label = target_names[ind] if fnd[ind]==0 else "")
                fnd[ind] = 1
    ax = plt.gca()
    min_x,max_x = ax.get_xlim()
    min_y,max_y = ax.get_ylim()
    x_arr = np.arange(min_x,max_x,0.1)
    y_arr = np.arange(min_y,max_y,0.1)
    x_coord,y_coord = np.meshgrid(x_arr,y_arr)
    all_points = np.c_[x_coord.ravel(),y_coord.ravel()]
    output = svc.decision_function(all_points).reshape(x_coord.shape)
    plt.contour(x_coord, y_coord, output, colors='k', levels=[-1, 0, 1],linestyles=['--', '-', '--'])
    plt.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=100,linewidth=1, facecolors='none', edgecolors='k',label = 'Support Vectors')
    plt.title(str(kernel)+" kernel; "+str(target_names[f1])+" , "+str(target_names[f2])+" ; "+class1+" , "+class2)
    plt.xlabel(class1)
    plt.ylabel(class2)
    plt.legend()
#     plt.savefig("q3 images/"+str(kernel)+" kernel with classes "+str(target_names[f1])+" and "+str(target_names[f2])+" between "+class1+" and "+class2)
    plt.show()
```

```python
for i in range(3):
    for j in range(3):
        if(i>=j):
            continue
        temp_df = pd.DataFrame(df.loc[df['target']!=3-i-j])
        temp_df.index = range(100)
        new_X = temp_df.drop(columns = ['target']).values
        new_y = temp_df['target'].values
        train_X,test_X,train_y,test_y = train_test_split(new_X,new_y,test_size = 0.3,shuffle = True)
        print("number of different classes: ",pd.DataFrame(train_y).nunique())
        for kernel in ['linear','rbf']:
            for C,plot_fig in zip([1],[True]):
                for l in range(2):
                    if(l==0):
                        solve(C,train_X[:,:2],train_y,test_X[:,:2],test_y,kernel,columns[0],columns[1],i,j,plot_fig)
                    else:
                        solve(C,train_X[:,2:],train_y,test_X[:,2:],test_y,kernel,columns[2],columns[3],i,j,plot_fig)
```

```python
def solve_without_fig(C,train_X,train_y,test_X,test_y,kernel,class1,class2):
    svc = SVC(kernel = kernel,C = C)
    svc.fit(train_X,train_y)
    pred = svc.predict(test_X)
    report = classification_report(test_y,pred,output_dict = True,target_names = target_names)
    report = pd.DataFrame(report).transpose()
    print("Using SVM with features "+class1+" and "+class2+ " ,kernel: "+kernel+ " and C  = "+str(C))
    display(report)
```

```python
for c in [0.001,1000]:
    new_X = df.drop(columns = ['target']).values
    new_y = df['target'].values
    train_X,test_X,train_y,test_y = train_test_split(new_X,new_y,test_size = 0.3,shuffle = True)
    for kernel in ['linear','rbf']:
        for i in range(2):
            if(i==0):
                solve_without_fig(c,train_X,train_y,test_X,test_y,kernel,columns[0],columns[1])
            else:
                solve_without_fig(c,train_X,train_y,test_X,test_y,kernel,columns[2],columns[3])
```
